```python
# Get the indices of all transmission lines in the network
lines_to_test = net.line.index
print(f"We will test N-1 contingencies for {len(lines_to_test)} lines.")
```

```python
# --- Reset and re-run OPF to get a clean network state ---
net_for_contingency = pn.case_ieee30()
pp.create_poly_cost(net_for_contingency, element=0, et='gen', cp1_eur_per_mw=10)
pp.create_poly_cost(net_for_contingency, element=1, et='gen', cp1_eur_per_mw=20)
# NOTE: In a real script, you'd add all costs again. We'll skip for brevity.
pp.runopp(net_for_contingency)


# --- Now, we run the same contingency loop ---
import pandas as pd

lines_to_test = net_for_contingency.line.index
contingency_results = []

for line_id in lines_to_test:
    # Create a copy of our clean, optimized network
    net_copy = net_for_contingency.copy()
    
    # Disconnect the line
    net_copy.line.in_service.loc[line_id] = False

    try:
        pp.runpp(net_copy)
    except pp.LoadflowNotConverged:
        status = "FAIL - Power flow did not converge"
        contingency_results.append({'contingency': f'Line {line_id}', 'status': status})
        continue

    # Check for overloads
    overloaded = net_copy.res_line[net_copy.res_line.loading_percent > 100]
    if not overloaded.empty:
        status = f"FAIL - {len(overloaded)} line(s) overloaded"
    else:
        status = "OK"
    
    contingency_results.append({'contingency': f'Line {line_id}', 'status': status})

results_df = pd.DataFrame(contingency_results)
print("--- Contingency Analysis Report ---")
print(results_df)
```

```python
import pandas as pd
import pandapower as pp
import pandapower.networks as pn
from copy import deepcopy # Correct import from Python's standard library

# 1. Start with a fresh network
net_for_contingency = pn.case_ieee30()

# 2. Run OPF to get our optimized starting point
pp.runopp(net_for_contingency)

# --- Run the contingency loop ---
lines_to_test = net_for_contingency.line.index
contingency_results = []

for line_id in lines_to_test:
    # Use the standard deepcopy function
    net_copy = deepcopy(net_for_contingency)
    
    net_copy.line.in_service.loc[line_id] = False

    try:
        pp.runpp(net_copy)
    except pp.LoadflowNotConverged:
        status = "FAIL - Power flow did not converge"
        contingency_results.append({'contingency': f'Line {line_id}', 'status': status})
        continue

    overloaded = net_copy.res_line[net_copy.res_line.loading_percent > 100]
    if not overloaded.empty:
        status = f"FAIL - {len(overloaded)} line(s) overloaded"
    else:
        status = "OK"
    
    contingency_results.append({'contingency': f'Line {line_id}', 'status': status})

results_df = pd.DataFrame(contingency_results)
print("--- Contingency Analysis Report ---")
print(results_df)
```
